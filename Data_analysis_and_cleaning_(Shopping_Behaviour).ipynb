```python
import pandas as pd

df = pd.read_csv(r'C:\Users\alwin\Downloads\customer_shopping_behavior.csv')
print(df.head())
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
#here every data is filled with a median of all 
#but we need to give categorically see median and fill accordingly
```

```python
df.isnull().sum()
```

```python
df.columns= df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']= pd.qcut(df['age'],q=4,labels=labels)
```

```python
df
```

```python
df[['age','age_group']].head(10)
```

```python
# creating column purchase_frequence_days

frequency_mapping={
    'Fortnightly':14 ,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14 ,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequence_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
